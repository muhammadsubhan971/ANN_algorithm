# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

# loading Dataset

```python
df=pd.read_csv("/content/heart_failure_clinical_records_dataset.csv")
df.head()
```

```python
df.info()
```

```python
df['DEATH_EVENT'].value_counts()
```

```python
x=df[df['DEATH_EVENT']==1]
y=df[df['DEATH_EVENT']==0]
```

```python
plt.scatter(x['age'],x['platelets'],color='red')
plt.scatter(y['age'],y['platelets'],color='blue')
# plt.show()
```

```python
X=df.drop('DEATH_EVENT',axis=1)
Y=df['DEATH_EVENT']
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
x_train
```

```python
sr=StandardScaler()
x_train=sr.fit_transform(x_train)
x_test=sr.transform(x_test)
```

```python
model=Sequential()
model.add(Dense(36,activation='relu',input_dim=12))
model.add(Dense(16,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='Adam')
```

```python
model.fit(x_train,y_train,epochs=300,validation_split=0.2)
```

```python
y_log=model.predict(x_test)
```

```python
y_pred=np.where(y_log>0.5,1,0)
```

```python
accuracy_score(y_test,y_pred)
```
